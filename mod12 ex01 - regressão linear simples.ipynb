### 1. Explorando a base de gorjetas

a. Na aula fizemos uma regressão de ```tip``` explicada por ```net_bill```. Carregue os pacotes e a base, crie as variáveis necessárias como na aula, e reproduza esta regressão.

b. Reproduza também o gráfico de dispersão dessas duas variáveis.

````python
# importando os pacotes
import pandas as pd
import seaborn as sns
from seaborn import load_dataset

import matplotlib.pyplot as plt
import numpy as np
import statsmodels.formula.api as smf
from sklearn.metrics import r2_score
````

````python
# baixando o dataframe tips e criando as colunas tip_pct e net_bill
tips = sns.load_dataset("tips")
tips["tip_pct"] = tips["tip"] / (tips["total_bill"] - tips["tip"])
tips["net_bill"] = tips["total_bill"] - tips["tip"]
tips.head()
````

````python
# gerando um grafico de dispersão
plt.figure(figsize=(10, 6))
sns.regplot( y = "tip", x = "net_bill", data = tips[tips["tip_pct"]<.5])
````

### 2. Mudança de perspectiva

Agora ajuste um modelo de regressão de ```tip_pct``` por ```net_bill```. Construa o gráfico de dispersão correspondente.

````python
#modelo de regressão de tip_pct por net_bill
reg = smf.ols('tip_pct ~ net_bill', data = tips).fit()
reg.summary()
````

````python
# gerando um grafico de dispersão
plt.figure(figsize=(10, 6))
sns.regplot( y = "tip_pct", x = "net_bill", data = tips[tips["tip_pct"]<.5])

# Adiciona a linha de regressão
y_pred = reg.predict()
plt.plot(tips['net_bill'], y_pred, color='red', label='Linha de Regressão', linewidth=2)

plt.title('Relação entre Percentual de Gorjeta e Conta Líquida')
plt.xlabel('Valor da Conta Líquida (net_bill)')
plt.ylabel('Percentual de Gorjeta (tip_pct)')
plt.legend()
plt.show()
````

### 4. Ajuste do modelo em outra escala

Com o modelo no item 3, obtenha a preditiva do valor da gorjeta:
  1. obtenha o valor predito ```pred``` para ```tip_pct``` (dica, veja o atributo ```reg.fittedvalues```)
  2. obtenha o valor predito para ```tip``` como ```tip_pct * net_bill```
  3. calcule o $R^2$ do modelo para ```tip```
  4. compare esse $R^2$ com o do item 1

````python
# 1. Obtendo o valor predito para tip_pct
pred = reg.fittedvalues
print("Valores preditos para tip_pct:\n", pred)
````

````python
# 2. Obtendo o valor predito para tip
tips['tip_predito'] = pred * tips['net_bill']
print("Valores preditos para tip:\n", tips['tip_predito'])
````

````python
# 3. Calculando o R^2 do modelo para tip
r2_tip = r2_score(tips['tip_predito'], tips['tip_pct'] * tips['net_bill'])
print(f"$R^2$ do modelo para tip: {r2_tip:.4f}")
````

````python
# 4. Calculando o R^2 do modelo para tip_pct
r2_tip_pct = reg.rsquared
print(f"$R^2$ do modelo para tip_pct: {r2_tip_pct:.4f}")
````

````python
# Comparando os R^2
if r2_tip > r2_tip_pct:
    print("O R^2 para tip é maior do que o R^2 para tip_pct.")
else:
    print("O R^2 para tip_pct é maior ou igual ao R^2 para tip.")
````
